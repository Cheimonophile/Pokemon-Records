```python
import pandas as pd
import plotly.express as px


levels = pd.read_csv('levels_table.csv')
levels = levels.set_index('Level')
levels = levels.drop(columns=(c for c in levels.columns if ".1" in c))
levels
```

```python
def find_exp(exp_type, level):
    return levels[exp_type][level]
find_exp('Fluctu­ating', 2)
```

```python
def find_level(exp_type, exp):
    return levels[levels[exp_type] <= exp].index.max()
find_level('Fluctu­ating', 4)
```

```python
HIGH_EXP_TYPE = 'Erratic'
LOW_EXP_TYPE = 'Fluctu­ating'
LEVEL_AT_NEW_POKEMON = 15

exp_diff = find_exp(HIGH_EXP_TYPE, LEVEL_AT_NEW_POKEMON)

for high_level in range(LEVEL_AT_NEW_POKEMON,101):
    high_exp = find_exp(HIGH_EXP_TYPE, high_level)
    low_exp = high_exp - exp_diff
    low_level = find_level(LOW_EXP_TYPE, low_exp)
    print(f'{high_level}\t{low_level}\t{"SAME LEVEL" if high_level <= low_level + 1 else ""}')

```
